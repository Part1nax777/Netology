```python
import pandas as pd

data = pd.read_csv('keywords.csv')
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
           'Северо-Запад': ['петербург', 'псков', 'мурманск'],
           'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
           }
def geolocation(row):
    for region in geo_data.keys(): 
        i = 0
        while i < len(geo_data[region]):
            if geo_data[region][i] in row.keyword:
                return region
            i += 1

data['region'] = data.apply(geolocation, axis=1).fillna('undefined')
```
