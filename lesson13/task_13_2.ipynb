```python
import requests
import pandas as pd
import json

EMAIL = ['xxx@x.ru', 'yyy@y.com']
leaked_emails = pd.DataFrame()
    
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
headers = {
    'Vaar-Header-App-Product': 'hackcheck-web-avast',
    'Vaar-Version': '0'
}
data = {
    'emailAddresses': EMAIL
}
    
page_object = requests.post(url, headers=headers, data=json.dumps(data))
```

```python
json_page_obj = json.loads(page_object.text)
for mails in json_page_obj['data'].values():
    mail = list(mails.keys())[0]
    breachId = str(mails[mail][0]['breachId'])
    site = json_page_obj['breaches'][breachId]['site']
    date = json_page_obj['breaches'][breachId]['publishDate']
    description = json_page_obj['breaches'][breachId]['description']
    row = {'Почта': mail, 'Дата утечки': date, 'Источник утечки': site, 'Описание утечки': description}
    leaked_emails = pd.concat([leaked_emails, pd.DataFrame([row])])

leaked_emails.drop_duplicates().reset_index().drop('index', 1)
```
