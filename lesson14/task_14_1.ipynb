* mean - среднее значение
* std - стандартное отклонение
* min - минимальное значение

- периверический пульс(категориальная): у большинства лошадей нормальное кровообращение.
- слизистые оболочки(категориальная): у 25% лошадей серьёзные проблемы с кровообращением.
