```python
import pandas as pd
new_data = pd.DataFrame()

data = pd.read_csv('horse_data.csv', na_values='?')
data.columns = ['операция?', 'Возраст', 'Номер больницы', 'Ректальная температура', 'Пульс', 'Частота дыхания',
              'Температура конечностей', 'Периферический пульс', 'Слизистые оболочки', 'Время наполнения капилляров',
              'Боль', 'Перистальтика', 'Вздутие живота', 'Назогастральный зонд', 'Назогастральный рефлюкс',
              'Назогастральный рефлюкс РН', 'Ректальное исследование', 'Живот', 'Объём упакованной ячейки', 'Общий белок',
              'Внешний вид брюшной полости', 'Общий белок брюшной полости', 'Результат', 'Хирургическое поражение?',
              'тип поражения_1', 'тип поражения_2', 'тип поражения_3', 'Паталогии']
pd.set_option('display.max_columns', None)
for column in data.columns:
    missing_values = data[column].isna().mean()
    print(f'{column} - {missing_values :.2%}')
```

* операция - значениями моды, т.к. всего один пропуск.
* Ректальная температура - можно подставить значение медианы, т.к. значение числовое особого разброса в данных нет
* Пульс - можно подставить значение медианы, т.к. значение числовое, особого разброса данных нет
* Частота дыхания - значениями моды, количество отсутствующих данных не велико, параметр не точный
* Температура конечностей - значениями моды, количество отсутствующих данных не велико
* Периферический пульс - значениями моды, параметр субьективный, значения основных данных после замены остаются в приемлимом диапазоне. 
* Слизистые оболочки - заменить на undefined, может сказаться на определении диагноза
* Время наполнения капиляров - заменить на значение медианы, значение достаточно относительное, могут быть ошибки при замерах.
* Боль - заменить на undefined, может повлиять на решение о проведении операции
* Перистальтика - заменить на undefined, может исказить вывод при определении диагноза
* Вздутие живота - заменить на undefined, важный параметр которые учитывается при решении о хирургическом вмешательстве
* Назогастральный зонд - заменить на значения моды, получение данных может быть не точным
* Назогастральный рефлюкс - заменить на значения моды, получение данных может быть не точным
* Назогастральный рефлюкс РН - заменить на undefined, слишком большое количество отсутствующих данных
* Ректальное исследование - заменить на undefined, достаточо большое количество отсутствующих значений, может повлиять на общий вывод
* Живот - заменить на undefined, достаточно большое количество отсутствующих значений, может повлиять на общий вывод.
* Объем упакованной ячейки - заменить значениями медианы, значение числовое, особого разброса нет (без учёта выбросов)
* Общий белок - заменить значениями медианы, значение числовое, особого разброса данных нет
* Внешний вид брюшной полости - undefined, может сказаться при постановке диагноза
* Общий белок брюшной полости - undefined, слишком большое количество отсутствующих данных
* Результат - заменить значениями моды, пропусков небольшое количество

```python
data['операция?'].fillna(data['операция?'].mode()[0], inplace = True)
data['Ректальная температура'].fillna(data['Ректальная температура'].median(), inplace = True)
data['Пульс'].fillna(data['Пульс'].median(), inplace = True)
data['Частота дыхания'].fillna(data['Частота дыхания'].mode()[0], inplace = True)
data['Температура конечностей'].fillna(data['Температура конечностей'].mode()[0], inplace = True)
data['Периферический пульс'].fillna(data['Периферический пульс'].mode()[0], inplace = True)
data['Слизистые оболочки'].fillna('undefined', inplace = True)
data['Время наполнения капилляров'].fillna(data['Время наполнения капилляров'].median(), inplace = True)
data['Боль'].fillna('undefined', inplace = True)
data['Перистальтика'].fillna('Перистальтика', inplace = True)
data['Вздутие живота'].fillna('Вздутие живота', inplace = True)
data['Назогастральный зонд'].fillna(data['Назогастральный зонд'].mode()[0], inplace = True)
data['Назогастральный рефлюкс'].fillna(data['Назогастральный рефлюкс'].mode()[0], inplace = True)
data['Назогастральный рефлюкс РН'].fillna('undefined', inplace = True)
data['Ректальное исследование'].fillna('undefined', inplace = True)
data['Живот'].fillna('undefined', inplace = True)
data['Объём упакованной ячейки'].fillna(data['Объём упакованной ячейки'].median(), inplace = True)
data['Общий белок'].fillna(data['Общий белок'].median(), inplace = True)
data['Внешний вид брюшной полости'].fillna('undefined', inplace = True)
data['Общий белок брюшной полости'].fillna('undefined', inplace = True)
data['Результат'].fillna(data['Результат'].mode()[0], inplace = True)

data.loc[:, data.columns.isin(['операция?', 'Ректальная температура', 'Пульс', 'Частота дыхания', 
                               'Температура конечностей', 'Периферический пульс', 'Время наполнения капилляров',
                              'Перистальтика', 'Вздутие живота', 'Назогастральный зонд', 'Назогастральный рефлюкс',
                              'Объём упакованной ячейки', 'Общий белок', 'Результат'])].describe()
```
