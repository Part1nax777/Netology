```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

data = pd.read_csv('googleplaystore.csv')
data_by_category = data.drop(data[data.Category == '1.9'].index) #удаление выброса
data_sum = list(data_by_category.Rating.groupby(data_by_category['Category']).sum()) 
data_count = list(data_by_category.Rating.groupby(data_by_category['Category']).count())
index = data_by_category.Category.unique()
plot_data = pd.DataFrame({'sum': data_sum, 'count': data_count}, index=index)
data_chart = plot_data.plot(kind = 'bar', title = 'Сумма оценок приложений googleplay в разрезе категорий', 
            xlabel = 'Категория', ylabel = 'Значение суммы')
plt.show()
```

Больше всего оценвают финансовые приложения(скорее всего это пользовательские приложения банковских систем), средняя оценка данных приложений около 4х баллов. Также часто оценивают приложения связанные с домашним бытом (возможно это связано с развитием IoT технологий).
