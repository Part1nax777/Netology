```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

data_2019 = pd.read_csv('2019.csv')
plt.plot(data_2019['Overall rank'].head(20), data_2019['Social support'].head(20), label = 'Социальная поддержка')
plt.plot(data_2019['Overall rank'].head(20), data_2019['GDP per capita'].head(20), label = 'ВВП на душу населения')
plt.plot(data_2019['Overall rank'].head(20), data_2019['Perceptions of corruption'].head(20), label = 'Восприятие коррупции')
plt.legend()
plt.title('Влияние экономических характеристик на "уровень счастья" страны')
plt.xlabel('Ранг страны')
plt.ylabel('Значение индекса')
plt.grid(which = 'minor', linestyle = ':')
plt.grid(which = 'major')
plt.minorticks_on()
plt.show()
```

В большей мере на "уровень счастья" страны вляет социальная поддержка населения и восприятие коррупции. В меньшей мере ВВП на душу населения.
