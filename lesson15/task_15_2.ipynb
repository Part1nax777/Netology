```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

data_2019 = pd.read_csv('2019.csv')
plt.plot(data_2019['Overall rank'].head(20), data_2019['Social support'].head(20), label = ' ')
plt.plot(data_2019['Overall rank'].head(20), data_2019['GDP per capita'].head(20), label = '   ')
plt.plot(data_2019['Overall rank'].head(20), data_2019['Perceptions of corruption'].head(20), label = ' ')
plt.legend()
plt.title('    " " ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.grid(which = 'minor', linestyle = ':')
plt.grid(which = 'major')
plt.minorticks_on()
plt.show()
```
