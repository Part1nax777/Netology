```python
import pandas as pd
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
data = pd.read_csv('winequality-red.csv')
group_data = data.groupby(data.quality).mean()
sns.heatmap(group_data)
```

```python
group_data_del_dioxide = group_data.drop(['free sulfur dioxide', 'total sulfur dioxide'], axis = 1)
sns.heatmap(group_data_del_dioxide)
```

```python
group_data_del_alcohol_fix_acidity = group_data_del_dioxide.drop(['alcohol', 'fixed acidity'], axis = 1)
sns.heatmap(group_data_del_alcohol_fix_acidity)
```

Ключевыми факторами влияющими на качество красного вина являются:
* остаточный сахар
* летучая кислотность
* критичная кислота
* фиксированная кислотность
* процент алкоголя
* общий диоксид серы
* свободный диоксид серы

Для более качественного вина данные показатели должны быть:
* остаточный сахар - 2,75 у.е.
* летучая кислотность - 0.8 у.е.
* критичная кислота - 0.75 у.е
* фиксированная кислотность - 8.5 у.е.
* процент алкоголя - 12 у.е.
* общий диоксид серы - 30 у.е.
* свободный диоксид серы - 20 у.е.
