```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

data = pd.read_csv('2019.csv')
columns = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices',
          'Generosity', 'Perceptions of corruption']
sns.pairplot(data[columns])
```

* При увеличении продолжительности жизни увеличивается ВВП на душу населения, социальная поддержка, свобода выбора жизненных интересов.
* При увеличении свободы выбора жизненных интересов увеличивается ВВП на душу населения, социальная поддержка, продолжительность жизни.
